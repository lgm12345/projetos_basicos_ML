```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
final_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = final_data["not.fully.paid"]
x = final_data.drop(columns=["not.fully.paid"],axis=1)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
predições = dtree.predict(x_test)
predições
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test,predições))
```

```python
confusion_matrix(y_test,predições)
```

```python
from sklearn.ensemble import RandomForestClassifier 
rand_forest = RandomForestClassifier(n_estimators=300)
rand_forest.fit(x_train,y_train)
```

```python
predições = rand_forest.predict(x_test)
predições
```

```python
classification_report(y_test,predições)
```

```python
confusion_matrix(y_test,predições)
```
