```python
#preciso analisar a correlaÃ§Ã£o entre as variÃ¡veis
#para analisar a correlaÃ§Ã£o entre as variaÃ©veis posso fazer grÃ¡ficos de regresÃ£o 
#linear para cada variavel com a que eu quero prever(se o cliente clicara ou nÃ£o)
y = ad_data["Clicked on Ad"]
x = ad_data[["Daily Time Spent on Site","Age","Area Income","Daily Internet Usage","Male"]]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from  sklearn.linear_model import LogisticRegression
reg_log = LogisticRegression()
reg_log.fit(x_train,y_train)
```

```python
prediÃ§Ãµes = reg_log.predict(x_test)
prediÃ§Ãµes
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,prediÃ§Ãµes))
```

```python
confusion_matrix(y_test,prediÃ§Ãµes)
```
