```python
from sklearn.model_selection import train_test_split
Y = clientes["Yearly Amount Spent"]
X = clientes[["Avg. Session Length","Time on App","Time on Website","Length of Membership","Yearly Amount Spent"]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
print(lm.coef_)
```

```python
print(lm.intercept_)
```

```python
predictions = lm.predict(x_test)
predictions
```

```python
sns.scatterplot(x=y_test,y=predictions)
from sklearn import metrics
```

```python
plt.scatter(y_test,predictions)
```

```python
print("MAe:" , metrics.mean_absolute_error(y_test,lm.predict(x_test)))
print("MSe:" , metrics.mean_squared_error(y_test,lm.predict(x_test)))
print("RMSe:" , np.sqrt(metrics.mean_squared_error(y_test,lm.predict(x_test))))
```

```python
#resÃ­duos:diferenca entre o valor previsto e o real
predictions = lm.predict(x_test)
sns.distplot((y_test - predictions), bins=30)
```
